```python
import astropy.coordinates as coord
from astropy.time import Time
import astropy.units as u
import time

def convert(t):
    return time.mktime(time.strptime(t, '%Y-%m-%d %H:%M:%S'))

def convert_deg(d):
    deg=d.split("d")[0]
    m=d.split("d")[1].split("m")[0]
    s=d.split("m")[1].split("s")[0]
    deg=float(deg)+float(m)/60+float(s)/3600
    return deg
```

```python
lat=46+36/60+54.9/3600
lon=15+10/60+19.5/3600

loc = coord.EarthLocation(lon=lon* u.deg,lat=lat * u.deg)
now = Time.now()
print(now,type(now))


'''dt=
max_az='''
df_hourly={}
df_hourly["date"]=[]
df_hourly["month"]=[]
df_hourly["day"]=[]
df_hourly["H"]=[]
df_hourly["M"]=[]
df_hourly["alt"]=[]
df_hourly["az"]=[]

df={}
df["date"]=[]
df["month"]=[]
df["max alt"]=[]
df["rise az"]=[]
df["rise"]=[]
df["set az"]=[]
df["set"]=[]
df["dt"]=[]
for month in range(1,13,1):
    for day in range(1,32):
        azi=[]
        sun_rise=None
        sun_set=None
        sun_rise_az=None
        sun_set_az=None
        for h in range(0,23):
            for m in range(0,60,1):
                try:
                    t="2022-%02d-%02dT%02d:%02d:00"%(month,day,h,m)
                    altaz = coord.AltAz(location=loc, obstime=t)
                    sun = coord.get_sun(Time(t, format='isot', scale='utc'))
                    y=sun.transform_to(altaz).alt
                    y2=sun.transform_to(altaz).az
                    azi.append(convert_deg(str(y)))
                    df_hourly["date"].append(t)
                    df_hourly["month"].append(month)
                    df_hourly["day"].append(day)
                    df_hourly["H"].append(h)
                    df_hourly["M"].append(m)
                    df_hourly["alt"].append(azi[-1])
                    df_hourly["az"].append(convert_deg(str(y2)))
                    if len(azi)>1:
                        if azi[len(azi)-1]*azi[len(azi)-2]<0 and sun_rise==None:
                            sun_rise=t.split("T")[1]
                            sun_rise_az=convert_deg(str(y2))
                        if azi[len(azi)-1]*azi[len(azi)-2]<0 and sun_rise!=None:
                            sun_set=t.split("T")[1]
                            sun_set_az=convert_deg(str(y2))
                    
                except:
                    pass
        if len(azi)>1:
            H=float(sun_set.split(":")[0])-float(sun_rise.split(":")[0])
            M=float(sun_set.split(":")[1])-float(sun_rise.split(":")[1])
            dt=H+M/60
            print(month,day,"%.2f"%max(azi),sun_rise,sun_set,"%.2f"%dt,"%.2f"%sun_rise_az,"%.2f"%sun_set_az)
            df["date"].append(t)
            df["month"].append(month)
            df["max alt"].append(max(azi))
            df["rise"].append(sun_rise)
            df["set"].append(sun_set)
            df["rise az"].append(sun_rise_az)
            df["set az"].append(sun_set_az)
            df["dt"].append(dt)
```
